# 1. Ley de Amdahl

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The speedup obtained by having:

- *P*: the percentage of the application that can run in parallel
- *N*: the number of processing units (CPUs, computing nodes, etc.)

$$S(P,N) = \frac{1}{(1-P)+\frac{P}{N}}$$

See [Wikipedia](https://www.wikiwand.com/en/Amdahl%27s_law)

```python
speedup = lambda P,N: 1/(1-P+P*1./N)

range_N = [2**i for i in range(0,15)]
range_P = [.5, .75,.9,.95]
x = np.r_[1,2,3.6,5,6,6.5,6,5.5,5.6,5.3,5.,4.3,4.,3.7,3.5]

plt.figure(figsize=(10,6))
for P in range_P:
    S = [speedup(P,N) for N in range_N]
    plt.plot(range(len(range_N)), S, label="%.2f"%P)

plt.plot(range(len(range_N)), x, lw=4, alpha=.5, color="black", label="real app")

plt.legend(title="percentage parallel")
plt.grid()
plt.xticks(range(len(range_N)), range_N);
plt.xlabel("number of processing units")
plt.ylabel("speedup")
plt.axhline(1, color="black", ls="--", alpha=.5)
plt.xlim(0,len(range_N)-1)
```
